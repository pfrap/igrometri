```python
soggiorno1=pd.read_csv("./sensor/A_-_Soggiorno_StartDate20230302_EndDate20230515_2055090163.CSV")
soggiorno2=pd.read_csv("./sensor/A_-_Soggiorno_StartDate20230303_EndDate20230607.CSV")
soggiorno3=pd.read_csv("./sensor/A_-_Soggiorno_StartDate20230325_EndDate20230627_28226642.CSV")
soggiorno4=pd.read_csv("./sensor/A_-_Soggiorno_StartDate20230611_EndDate20230910.CSV")
soggiorno5=pd.read_csv("./sensor/A_-_Soggiorno_StartDate20230703_EndDate20231010_2085698185.CSV")

cucina1=pd.read_csv("./sensor/B_-_Cucina_StartDate20230302_EndDate20230515_666149429.CSV")
cucina2=pd.read_csv("./sensor/B_-_Cucina_StartDate20230303_EndDate20230607.CSV")
cucina3=pd.read_csv("./sensor/B_-_Cucina_StartDate20230325_EndDate20230627_343446979.CSV")
cucina4=pd.read_csv("./sensor/B_-_Cucina_StartDate20230611_EndDate20230910.CSV")
cucina5=pd.read_csv("./sensor/B_-_Cucina_StartDate20230703_EndDate20231010_2849756145.CSV")

balcone_est1=pd.read_csv("./sensor/C_-_balcone_est_StartDate20230302_EndDate20230515_982370467.CSV")
balcone_est2=pd.read_csv("./sensor/C_-_balcone_est_StartDate20230303_EndDate20230607.CSV")
balcone_est3=pd.read_csv("./sensor/C_-_balcone_est_StartDate20230325_EndDate20230627_214114977.CSV")
balcone_est4=pd.read_csv("./sensor/C_-_balcone_est_StartDate20230611_EndDate20230729.CSV")
balcone_est5=pd.read_csv("./sensor/C_-_balcone_est_StartDate20230703_EndDate20231010_1522464121.CSV")

balcone_nord1=pd.read_csv("./sensor/D_-_balcone_nord_StartDate20230302_EndDate20230515_3951463036.CSV")
balcone_nord2=pd.read_csv("./sensor/D_-_balcone_nord_StartDate20230303_EndDate20230607.CSV")
balcone_nord3=pd.read_csv("./sensor/D_-_balcone_nord_StartDate20230325_EndDate20230627_2193993605.CSV")
balcone_nord4=pd.read_csv("./sensor/D_-_balcone_nord_StartDate20230611_EndDate20230829.CSV")
balcone_nord5=pd.read_csv("./sensor/D_-_balcone_nord_StartDate20230703_EndDate20231010_1477550423.CSV")

bagno1=pd.read_csv("./sensor/E_-_Bagno_StartDate20230302_EndDate20230515_1870920116.CSV")
bagno2=pd.read_csv("./sensor/E_-_Bagno_StartDate20230325_EndDate20230627_1812495436.CSV")
bagno3=pd.read_csv("./sensor/E_-_Bagno_StartDate20230325_EndDate20230628.CSV")
bagno4=pd.read_csv("./sensor/E_-_Bagno_StartDate20230703_EndDate20231010_3306638433.CSV")
bagno5=pd.read_csv("./sensor/E_-_bagno_StartDate20230302_EndDate20230320.CSV")
bagno6=pd.read_csv("./sensor/E_-_bagno_StartDate20230322_EndDate20230325.CSV")

cameraAgne1=pd.read_csv("./sensor/White__camera_agne_StartDate20230302_EndDate20230515_718013837.CSV")
cameraAgne2=pd.read_csv("./sensor/White__camera_agne_StartDate20230302_EndDate20230607.CSV")
cameraAgne3=pd.read_csv("./sensor/White__camera_agne_StartDate20230325_EndDate20230627_303127903.CSV")
cameraAgne4=pd.read_csv("./sensor/White__camera_agne_StartDate20230610_EndDate20230910.CSV")
cameraAgne5=pd.read_csv("./sensor/White__camera_agne_StartDate20230703_EndDate20231010_3232901627.CSV")
```

# Merge dei vari dataframe e assegnazione nomi stanze

```python
data_frames_soggiorno = [soggiorno1,soggiorno2,soggiorno3,soggiorno4,soggiorno5]
data_frames_cucina = [cucina1, cucina2, cucina3, cucina4, cucina5]
data_frames_best = [balcone_est1,balcone_est2,balcone_est3,balcone_est4,balcone_est5]
data_frames_bnord = [balcone_nord1,balcone_nord2,balcone_nord3,balcone_nord4,balcone_nord5]
data_frames_bagno = [bagno1,bagno2,bagno3,bagno4,bagno5,bagno6]
data_frames_cameraAgne = [cameraAgne1,cameraAgne2,cameraAgne3,cameraAgne4,cameraAgne5]
```

```python
df_soggiorno = reduce(lambda  left,right: pd.merge(left,right,on=['Date', 'Temp', 'Umidità','Remark'], how='outer'), data_frames_soggiorno)
df_soggiorno["Stanza"]="Soggiorno"
df_soggiorno=df_soggiorno[['Stanza','Date', 'Temp', 'Umidità']]
```

```python
df_cucina = reduce(lambda  left,right: pd.merge(left,right,on=['Date', 'Temp', 'Umidità','Remark'], how='outer'), data_frames_cucina)
df_cucina["Stanza"]="Cucina"
df_cucina=df_cucina[['Stanza','Date', 'Temp', 'Umidità']]
```

```python
df_best = reduce(lambda  left,right: pd.merge(left,right,on=['Date', 'Temp', 'Umidità','Remark'], how='outer'), data_frames_best)
df_best["Stanza"]="Balcone est"
df_best=df_best[['Stanza','Date', 'Temp', 'Umidità']]
```

```python
df_bnord = reduce(lambda  left,right: pd.merge(left,right,on=['Date', 'Temp', 'Umidità','Remark'], how='outer'), data_frames_bnord)
df_bnord["Stanza"]="Balcone nord"
df_bnord=df_bnord[['Stanza','Date', 'Temp', 'Umidità']]
```

```python
df_bagno = reduce(lambda  left,right: pd.merge(left,right,on=['Date', 'Temp', 'Umidità','Remark'], how='outer'), data_frames_bagno)
df_bagno["Stanza"]="Bagno"
df_bagno=df_bagno[['Stanza','Date', 'Temp', 'Umidità']]
```

```python
df_cameraAgne = reduce(lambda  left,right: pd.merge(left,right,on=['Date', 'Temp', 'Umidità','Remark'], how='outer'), data_frames_cameraAgne)
df_cameraAgne["Stanza"]="Camera Agne"
df_cameraAgne=df_cameraAgne[['Stanza','Date', 'Temp', 'Umidità']]
```

```python
data_frames = [df_soggiorno,df_cucina,df_best,df_bnord,df_bagno,df_cameraAgne]
```

```python
df_totale = reduce(lambda  left,right: pd.merge(left,right,on=['Stanza','Date', 'Temp', 'Umidità'], how='outer'), data_frames)
```

# Modifica dtypes delle colonne

```python
df_totale["Temp"]=df_totale["Temp"].str.replace("℃","")
df_totale["Umidità"]=df_totale["Umidità"].str.replace("%","")
```

```python
df_totale["Stanza"]=df_totale["Stanza"].astype(str)
df_totale["Temp"]=df_totale["Temp"].astype(float)
df_totale["Umidità"]=df_totale["Umidità"].astype(float)
```

```python
df_totale["Date"]=pd.to_datetime(df_totale["Date"])
```

```python
df_totale.dtypes
```

```python
df_totale.query("Stanza=='Cucina'")
```

```python
df_totale.sort_values(by="Date", inplace=True)
```

# Andamento temperatura delle differenti stanze

```python
fig, ax=plt.subplots(figsize=(17,8), layout="constrained")
fig.suptitle('Temperatura °C - daily running average')

ax.plot(df_totale[df_totale["Stanza"]=="Cucina"].Date, df_totale[df_totale["Stanza"]=="Cucina"].Temp.rolling(window=144).mean(),label='Cucina', linewidth=1)
ax.plot(df_totale[df_totale["Stanza"]=="Soggiorno"].Date, df_totale[df_totale["Stanza"]=="Soggiorno"].Temp.rolling(window=144).mean(),label='Soggiorno',linewidth=1)
ax.plot(df_totale[df_totale["Stanza"]=="Bagno"].Date, df_totale[df_totale["Stanza"]=="Bagno"].Temp.rolling(window=144).mean(),label='Bagno',linewidth=1)
ax.plot(df_totale[df_totale["Stanza"]=="Camera Agne"].Date, df_totale[df_totale["Stanza"]=="Camera Agne"].Temp.rolling(window=144).mean(),label='Camera da letto',linewidth=1)
ax.plot(df_totale[df_totale["Stanza"]=="Balcone est"].Date, df_totale[df_totale["Stanza"]=="Balcone est"].Temp.rolling(window=144).mean(),label='Balcone est',linewidth=1)
ax.plot(df_totale[df_totale["Stanza"]=="Balcone nord"].Date, df_totale[df_totale["Stanza"]=="Balcone nord"].Temp.rolling(window=144).mean(), label='Balcone nord',linewidth=1)
plt.legend()
plt.savefig("./grafici/Temperatura °C - daily running average.jpg")
plt.show()
```
